Q1  Define the role of a Data Analyst in your own words. What value do they bring to an organization?

-   A Data Analyst looks at data to find useful information that helps a company make better decisions. They collect, clean, and study data, then create simple reports or charts to show trends and patterns.  
They help the business understand what’s working, spot problems, and make smarter decisions based on facts instead of guesses.

Q3   Write the end-to-end analytics workflow in the correct order and briefly explain each step.

-   The correct of end to end analyctics workflow are in orders:-
1. Define the Problem:   Decide what question you want to answer or what decision you want to support.  
2. Collect Data:    Gather data from sources like databases, surveys, or spreadsheets.  
3. Clean Data:   Fix mistakes, remove duplicates, and handle missing information so data is accurate.  
4. Analyze Data:   Look for patterns, trends, or relationships using tools like Excel, SQL, or Python.  
5. Visualize Data:    Make charts, graphs, or dashboards to explain findings clearly.  
6. Share Insights:  Present your results and recommendations to help the team or management make decisions.  
7. Take Action:  Use insights to improve decisions, then track results to see if changes worked.  

      This workflow helps turn raw data into useful insights for better business decisions.  

Q4  What is Prompt Engineering, and explain any two types of prompts used in Generative AI?

  -  Prompt Engineering is the process of writing clear instructions for  AI models so they give the best possible output. It like asking the AI the right question in the right way.  

Mainly two types of prompts used in Generative AI:  

1. Instruction based prompts: You tell the AI exactly what to do.  
   - Example: “Write a short story about a talking dog.”  

2. Few-shot prompts:  You give the AI a few examples so it can understand the pattern before generating its own output.

Q8  Explain the evolution of analytics, highlighting the different stages from Descriptive to Generative AI.

-1. Descriptive Analytics: It tells us what happened in the past by looking at data.  
   - Example: A report showing last months sales.  

2. Diagnostic Analytics: It explains why something happened by digging deeper into data.  
   - Example: Finding out why sales dropped last month.  

3. Predictive Analytics : It uses data to guess what might happen in the future.  
   - Example: Predicting next months sales based on trends.  

4. Prescriptive Analytics: It suggests what actions to take to get the best results.  
   - Example: Recommending which products to promote to increase sales.  

5. Generative AI: It creates new content or ideas based on data and learning.  
   - Example: AI writing emails, creating art, or generating reports automatically.
